```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("MAE_data.csv")
```

```python
df
```

```python
plt.rcParams["figure.figsize"] = [10.00,3.00]
plt.ylabel("MAE (knots)")
plt.scatter(df['x'],df['y1'],color = 'orange',marker = "o",alpha = 1)
plt.plot(df['x'],df['y1'],color = 'orange',alpha = 1,label = "Linear Regression")
plt.scatter(df['x'],df['y4'],color = 'blue',marker = "s",alpha = 1)
plt.plot(df['x'],df['y4'],color = 'blue',alpha = 1,label = "MLP")
plt.scatter(df['x'],df['y2'],color = 'green',marker = "s",alpha = 1)
plt.plot(df['x'],df['y2'],color = 'green',alpha = 1,label = "lasso")
plt.scatter(df['x'],df['y3'],color = 'red',marker = "s",alpha = 1)
plt.plot(df['x'],df['y3'],color = 'red',alpha = 1,label = "random forest")
xticks = [i for i in range(2010,2019)]
yticks = [7,8,9,10,11,12,13,14,15]
xtick_labels = [i for i in range(2010,2019)]
ytick_labels = [7,8,9,10,11,12,13,14,15]
plt.xticks(xticks,xtick_labels)
plt.yticks(yticks,ytick_labels)
plt.grid(True,color = 'black',alpha = 0.25)
plt.legend(loc = "upper center")
plt.text(2011,13.5,"(c) MAE",fontsize = 15)
plt.show()
```
